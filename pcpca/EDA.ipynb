```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
states = pd.read_csv("covid_12h.csv")
```

# Highest Variance Features

```python
columns = ['ALBUMIN', 'ANION GAP',
       'BASE EXCESS', 'BICARBONATE', 'BILIRUBIN', 'CALCIUM',
       'CARBOXYHEMOGLOBIN', 'CHLORIDE', 'CREATININE', 'HEMATOCRIT',
       'HEMOGLOBIN', 'INSPIRED OXYGEN', 'INTERNATIONAL NORMALIZED RATIO',
       'LACTATE', 'METHEMOGLOBIN', 'PARTIAL THROMBOPLASTIN TIME', 'PCO2', 'PH',
       'PLATELETS', 'PO2', 'POTASSIUM', 'SODIUM', 'UREA NITROGEN',
       'URINE OUTPUT', 'WHITE BLOOD CELLS', 'FIO2', 'PEEP', 'OXYGEN (L)',
       'Respiratory Aids', 'Nonsteroidal Anti-inflammatory Agents (NSAIDs)',
       'Corticosteroids - Topical', 'Mineralocorticoids',
       'Glucocorticosteroids', 'Influenza Agents', 'Antiretrovirals']
```

```python
for feature in columns:
    positive = states[states[feature] >= 0]
    normal = positive[states[feature] < 99999]
    values = list(normal[feature])
    #sns.histplot(values, bins=100)
    print(str(feature) + ": " + str(len(values)) + " non-negative values, Variance: " + str(np.var(values)))
```

```python
highvar_features = ['CHLORIDE', 'HEMATOCRIT',
       'INSPIRED OXYGEN', 
       'LACTATE', 'PARTIAL THROMBOPLASTIN TIME', 'PCO2', 
       'PLATELETS', 'PO2', 'SODIUM', 'UREA NITROGEN',
       'URINE OUTPUT', 'WHITE BLOOD CELLS', 'FIO2', 'OXYGEN (L)']
```

```python
for f in highvar_features:
    positive = states[states[f] >= 0]
    normal = positive[states[f] < 99999]
    values = list(normal[f])
    plt.title("Histogram of " + str(f))
    sns.distplot(values)
    plt.show()
    plt.close()
```
